```python
import matplotlib.pyplot as plt
import datetime
import matplotlib.dates as mdates

# Dfinir les dates de dbut et de fin de chaque tche
start_dates = [datetime.datetime(2023, 10, 1), datetime.datetime(2023, 10, 1), datetime.datetime(2023, 10, 1), 
               datetime.datetime(2023, 11, 1), datetime.datetime(2023, 12, 1), datetime.datetime(2023, 12, 15), 
               datetime.datetime(2024, 1, 1), datetime.datetime(2024, 3, 1), datetime.datetime(2024, 3, 15), 
               datetime.datetime(2024, 4, 1)]
end_dates = [datetime.datetime(2023, 10, 31), datetime.datetime(2024, 4, 15), datetime.datetime(2023, 11, 30),
             datetime.datetime(2023, 11, 30), datetime.datetime(2023, 12, 14), datetime.datetime(2023, 12, 31),
             datetime.datetime(2024, 2, 28), datetime.datetime(2024, 3, 14), datetime.datetime(2024, 4, 30), 
             datetime.datetime(2024, 5, 15)]

# Dfinir le nom de chaque tche
tasks = ["Demander l'accs  la base de donnes", 
         "Explorer les donnes", 
         "Achever la lecture de la thse de Michel Trigalet", 
         "Dbuter la rdaction", 
         "Prvoir un rendez-vous avec le directeur de mmoire", 
         "Faire un rapport  notre directeur", 
         "Achever la rdaction jusqu' la mthodologie",
         "Faire un rapport notre directeur sur notre avance",
         "Achever la rdaction",
         "Envoyer le mmoire"]

# Crer le diagramme de Gantt
plt.figure(figsize=(10, 8))
for i in range(len(tasks)):
    plt.barh(i, end_dates[i] - start_dates[i], left=start_dates[i], height=0.3, align='center', alpha=0.8)
plt.ylim(-0.5, len(tasks)-0.5)
plt.yticks(range(len(tasks)), tasks)
plt.gca().invert_yaxis()  # inverser l'ordre des tches
plt.gca().xaxis_date()  # rendre l'axe x compatible avec les dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # formater l'axe x avec les mois
plt.grid(axis='x')  # ajouter une grille sur l'axe x
plt.title('Diagramme de Gantt')
plt.show()
```
